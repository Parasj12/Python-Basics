# Data manipulations 1

```python
import pandas as pd
```

```python
cust_data=pd.read_csv("Cust_data.csv")
```

```python
cust_demo=pd.read_csv("Cust_demo.csv")
cust_demo.head()
```

1. Create sub set of cust_demo as “cust_s1” with ID, age, Gender and Location variables using the condition marital_status = “Married” and Own_house = 1 and age>28.

```python
cust_s1=cust_demo[["ID","age","Gender","Location"]][(cust_demo["age"]>28) & (cust_demo["Martial_Status"]=="Married") & (cust_demo["Own_House"]==1)]
cust_s1.head()
```

2. Sort the cust_s1 dataset using location in ascending order and age in descending order with in the location.

```python
cust_s1.sort_values(by=['Location','age'],ascending=[True,False],inplace=True)
cust_s1.head()
```

3. Rename the variable “NumberOfDependents” as “No_of_dependents” in cust_demo data.

```python
cust_demo=cust_demo.rename(columns={'NumberOfDependents':'No_of_dependents'})
cust_demo.head()
```

4. Remove duplicates from the both data sets Cust_data & Cust_Demo.


```python
cust_data=cust_data.drop_duplicates()
cust_data.head()
```

```python
cust_demo=cust_demo.drop_duplicates()
cust_demo.head()
```

5. Create new variable as “No_of_30_plus_DPD = No_of_30_59_DPD+No_of_90_DPD+ No_of_60_89_DPD”, “No_of_60_plus_DPD = No_of_90_DPD+ No_of_60_89_DPD” in cust_data.

```python
cust_data['No_of_30_plus_DPD']=cust_data['No_of_30_59_DPD']+cust_data['No_of_90_DPD']+cust_data['No_of_60_89_DPD']
cust_data['No_of_60_plus_DPD']=cust_data['No_of_90_DPD']+cust_data['No_of_60_89_DPD']             
cust_data.head()
```

6. Perform below joins between cust_data & cust_demo.
<br>
a. Create data set “cust_leftjoin” using Left Join
<br>
b. Create data set “cust_rightjoin” using right Join
<br>c. Create data set “cust_innterjoin” using inner Join
<br>d. Create data set “cust_fulljoin” using full Join


```python
cust_leftjoin=cust_data.join(cust_demo.set_index('ID'),lsuffix='_l',rsuffix='_r')
cust_leftjoin.head()
```

```python
cust_rightjoin=cust_data.join(cust_demo.set_index('ID'),how='right',lsuffix='_l',rsuffix='_r')
cust_rightjoin.head()
```

```python
cust_innerjoin=cust_data.join(cust_demo.set_index('ID'),how='inner',lsuffix='_l',rsuffix='_r')
cust_innerjoin.head()
```

```python
cust_fulljoin=cust_data.join(cust_demo.set_index('ID'),how='outer',lsuffix='_l',rsuffix='_r')
cust_fulljoin.head()
```

7. Create two data sets by taking random sample from cust_demo with 5% data in first data set and 10000 observations in second data set.


```python
cust_sample1=cust_demo.sample(frac=0.05,random_state=13)
cust_sample1.head()
```

```python
cust_sample2=cust_demo.sample(n=10000,random_state=13)
cust_sample2.head()
```

8. Calculate “number of customers” and “number of married customer” using combined data set comes from left join of cust_data and cust_demo


```python
numberofcustomers=cust_leftjoin['ID'].count()
numberofcustomers
```

```python
numberofmarriedcustomers=cust_leftjoin['Martial_Status'][cust_leftjoin.Martial_Status=="Married"].count()
numberofmarriedcustomers
```
