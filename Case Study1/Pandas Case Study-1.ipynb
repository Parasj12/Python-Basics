```python
#6. In customer table, create a new variable called “customer value segment” using customer value as follows. -  
#- High Value Segment - > 25000  
#- Medium Value Segment – Between 10000 and 25000  
#- Low Value Segment – less than or equal to 10000 
```

```python
max=cust['Customer_Value'].max()
```

```python
bin=[0,10000,25000,max]
label=['Low Value Segment','Medium Value Segment','High Value Segment']
cust['Customer_Value_Segment']=pd.cut(cust['Customer_Value'],bins=bin,labels=label)

cust.head()
```

```python
#7. Create variables “average revenue per trip” and “balance points” in the customer data set.  
```

```python
cust['Balance points']=cust['Points_Earned']-cust['Points_Redeemed']
```

```python
cust['Average_Revenue_per_trip']=cust['Customer_Value']/cust['Balance points']
cust.head()
```

```python
#8. How many days between last purchase date and today?
```

```python
import datetime as dt
today=pd.to_datetime('today')
today
```

```python
today=pd.to_datetime(today,format=("%Y%m%d")).date()
```

```python
def dat(col):
    col=pd.to_datetime(col,format=("%Y%m%d")).date()
    return col
```

```python
cust['Recent_Date']=cust['Recent_Date'].apply(dat)
```

```python
cust['No_of_Days']=today-cust['Recent_Date']
```
